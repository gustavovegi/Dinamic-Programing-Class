```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#Matriz Rotao
#definio das direes de olhar do personagem (em graus)
direcoes = [0, 45, 90, 135, 180]

#Funo para aplicar a rotao em um ponto (usando uma matriz de rotao)
def aplicar_rotacao(ponto, angulo):
    theta = np.radians(angulo)
    matriz_rotacao = np.array([[np.cos(theta), -np.sin(theta)],
                               [np.sin(theta), np.cos(theta)]])
    return np.dot(matriz_rotacao, ponto)

#Plotagem das diferentes direes de olhar do personagem
plt.figure(figsize=(10, 5))
plt.title('Animao: Personagem olhando em diferentes direes')

#Para cada direo, plotamos a posio do personagem aps a rota
for angulo in direcoes:
    #definindo a posio inicial do personagem (em diferentes pontos do plano)
    posicao_inicial = np.array([2* np.cos(np.radians(angulo)), 2 * np.sin(np.radians(angulo))])
    posicao_rotacionada = aplicar_rotacao(posicao_inicial, angulo)
    plt.scatter(posicao_rotacionada[0], posicao_rotacionada[1], label=f'direo{angulo}', marker='o', s=200)

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```

```python
#Matriz de Escala 

#definio dos pontos dos objetos na cena
moeda = np.array([[0, 0], [1, 0], [1, 1], [0, 1], [0, 0]]) #pontos da moeda(um quadrado)
obstaculo = np.array([[0, 0], [2, 0], [2, 1], [0, 1], [0, 0]]) #pontos do obstaculo (um retangulo)

#funo para aplicar a matriz de escala em um conjunto de pontos
def aplicar_escala(pontos, fator_x, fator_y):
        matriz_escala = np.array([[fator_x, 0],
                                  [0, fator_y]])
        return np.dot(pontos, matriz_escala)

#plotagem dos objetos antes e depois do redimensionamento
plt.figure(figsize=(10, 5))
plt.title('Redimensionamento de Objetos')

#plotagem da moeda antes do redimensionamento
plt.plot(moeda[:, 0], moeda[:, 1], label='Moeda (antes)', color='blue')

#aplicao da matriz de escala para redimensionar a moeda 
moeda_redimensionada = aplicar_escala(moeda, 1.5, 1.5)
plt.plot(moeda_redimensionada[:, 0], moeda_redimensionada[:, 1], label='Moeda (depois)', linestyle='dashed', color='blue')

#plotagem do obstaculo antes do redimensionamento
plt.plot(obstaculo[:, 0], obstaculo[:, 1], label='Obstaculo (antes)', color='red')

#aplicao da matriz de escala para redimensionar o obstaculo
obstaculo_redimensionado = aplicar_escala(obstaculo, 0.5, 0.5)
plt.plot(obstaculo_redimensionado[:, 0], obstaculo_redimensionado[:, 1], label='Obstaculo (depois)', linestyle='dashed', color='red')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```

```python
#Matriz de translao
#definio do ponto inicial (x, y)
ponto =  np.array([1, 1])
 
#funo para aplicar a matriz de translao em um ponto
def transladar_ponto(ponto, dx, dy):
    #criar matriz de translao 
    matriz_translacao = np.array([[1, 0, dx],
                                  [0, 1, dy], 
                                  [0, 0, 1]])
    
    #adicionar uma dimenso ao ponto para que seja uma matriz 3x1
    ponto_homogeneo = np.hstack((ponto, 1))

    #aplicar a matriz de translao ao ponto
    ponto_transformado = np.dot(matriz_translacao, ponto_homogeneo)

    #retornar as coordenadas x e y do ponto transformado
    return ponto_transformado[:2]

#nova posio para transladar o ponto
nova_posicao = (2, 3)

#transladar o ponto para a nova posio 
ponto_transformado = transladar_ponto(ponto, *nova_posicao)

#plotar o ponto original e o ponto transladado
plt.figure(figsize=(6, 6))
plt.plot(ponto[0], ponto[1], 'bo', label='Ponto Original')
plt.plot(ponto_transformado[0], ponto_transformado[1], 'ro', label='ponto transladado')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Exemplo de matriz de translao')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```

```python
#contagem de palavras 
def contar_palavras(lista_palavras):
    contagem={}
    for palavra in lista_palavras:
        contagem[palavra] = contagem.get(palavra, 0) + 1
    return contagem

lista = ['a', 'a', 'c', 'd', 'e', 'r', 'e']
print(contar_palavras(lista))
```

```python
#media notas

def calcular_media_notas(notas_alunos):
    medias= {}
    for aluno, notas in notas_alunos.items():
        media = sum(notas) / len(notas)
        medias[aluno] = media
    return medias

notas_alunos = {'Joo': [8, 9, 10], 'Aboubakar': [0, 0, 0], 'Pedro': [10, 10, 10]}
print(calcular_media_notas(notas_alunos))
```

```python
#Busca de produtos 

def busca_produto(codigo, dicionario_produtos):
    return dicionario_produtos.get(codigo, "Produto no encontrado")

produtos = {'001':'Camiseta', '002': 'Cala', '003': 'Meia'}
print(busca_produto('002', produtos))
```

```python
#contagem de letras

def contar_letras(frase):
    contagem = {}
    for letra in frase:
        if letra.isalpha():
            contagem[letra] = contagem.get (letra, 0) + 1
    return contagem

frase = "Hello, nigga"
print(contar_letras(frase))
```

```python
def juntar_dicionarios(dic1, dic2):
    novo_dic = dic1.copy()
    for chave, valor in dic2.items():
        novo_dic[chave] = novo_dic.get(chave, 0) + valor
    return novo_dic

dic1={'a': 1, 'b': 2}
dic2={'b': 3, 'c': 4}

juntar = juntar_dicionarios(dic1, dic2)
juntar
```
